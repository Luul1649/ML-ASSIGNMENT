```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = r"C:\Users\hi\Downloads\diabetes.csv"  # Update path if needed
df = pd.read_csv(file_path)

# Set the style
sns.set(style="whitegrid")

# Define the target and feature columns
target_variable = "Outcome"
features = df.columns[df.columns != target_variable]

# Plot each independent variable against the outcome
for feature in features:
    plt.figure(figsize=(6, 4))
    sns.boxplot(x=target_variable, y=feature, data=df)
    plt.title(f"{feature} vs {target_variable}")
    plt.tight_layout()
    plt.show()
```
